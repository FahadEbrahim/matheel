## Guided pairwise check

After the app opens, use the **Pairwise** tab for a quick check that does not download a model:

1. Paste `def add(a, b):\n    return a + b` into **Code A**.
2. Paste `def add(x, y):\n    return x + y` into **Code B**.
3. In **Metrics**, set **Metric Preset** to **Lexical Only**.
4. Select **Run Pair** and inspect the overall score and metric breakdown.

The lexical preset keeps this first run fast. Semantic presets can download model weights and take longer on their first run.
